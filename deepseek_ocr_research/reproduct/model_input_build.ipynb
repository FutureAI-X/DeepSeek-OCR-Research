```python
# 导入依赖包
from image_handle import load_image_pil_rgb, BasicImageTransform
from PIL import ImageOps
import torch
```

```python
# 定义输入变量
# 用户输入的问题, 简单起见, 固定为这个值, 图像会放在 prompt 之前
prompt="\nExtract the text in the image."
# 图片路径
image_file = 'tengwangge.png'
# 图片中的文本(训练时使用)
image_text = ''
```

# 一 图像处理

```python
# 图片的全局视图列表, 元素为张量, 每个元素的shape为[C, H, W]
images_list = []
```

```python
# 加载原始图像, 以 RGB 模式加载
image = load_image_pil_rgb(image_file)
```

```python
# 将图像调整为统一大小(通过resize, pad操作)
image = ImageOps.pad(image, (640, 640), color=tuple([127, 127, 127]))
```

```python
# 转换为 tensor, 形状为[C, H, W]即[通道数, 高度, 宽度], 对于 RGB 图像, 通道数应该为 3
image_transform=BasicImageTransform(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5), normalize=True)
image = image_transform(image).to(torch.bfloat16)
image.shape
```

```python
# 将图片添加到全局视图列表中
images_list.append(image)
```
